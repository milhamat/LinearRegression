```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
3.29e7
```

```python
X = DataFrame(data, columns=['production_budget_usd']) 
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y, alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross')
plt.ylim(0,3e9)
plt.xlim(0,45e7)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X,y)
```

Slope Coeffision :

```python
regression.coef_ # theta_1
# regression coefficient is value represents the mean change in the response given a one unit change in the predictor
# jadi angka 3.115 itu angka pertambahan rata"
```

```python
# Intercept

regression.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y, alpha=0.3)
plt.plot(X, regression.predict(X), color='green', linewidth=3)

plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross')
plt.ylim(0,3e9)
plt.xlim(0,45e7)
plt.show()
```

```python
regression.score(X,y)
```
